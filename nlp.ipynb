```python
import pandas as pd

df=pd.read_csv('fake_job_postings.csv')
df.head()
df.info()
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text_cols = ['title','description','requirements','benefits','company_profile']

df['text'] = df[text_cols].fillna('').agg(' '.join, axis=1)
```

```python
df['text'].head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

corpus = []

for i in range(len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['text'][i])
    review = review.lower()
    review = review.split()

    review = [lemmatizer.lemmatize(word) for word in review if word not in stop_words]

    review = " ".join(review)
    corpus.append(review)
```

```python
len(corpus)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=10000)
X = tv.fit_transform(corpus).toarray()
```

```python
df[['min_salary','max_salary']] = df['salary_range'].str.split('-', expand=True)

df['min_salary'] = pd.to_numeric(df['min_salary'], errors='coerce')
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce')

df['avg_salary'] = (df['min_salary'] + df['max_salary']) / 2

df['avg_salary'] = df['avg_salary'].fillna(0)

df.drop(['salary_range','min_salary','max_salary'], axis=1, inplace=True)
```

```python
df.drop(['job_id','title','description','requirements','benefits','company_profile'], axis=1, inplace=True)
df.head()
```

```python
# split location into country, state, city
loc = df['location'].str.split(',', n=2, expand=True)

df['country'] = loc[0]
df['state'] = loc[1]
df['city'] = loc[2]

# fill missing values
df[['country','state','city']] = df[['country','state','city']].fillna('unknown')

# encode
from sklearn.preprocessing import LabelEncoder

for col in ['country','state','city']:
    df[col] = LabelEncoder().fit_transform(df[col])

# drop original column
df.drop('location', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

cols = ['department','employment_type','required_experience',
        'required_education','industry','function']

le = LabelEncoder()

for col in cols:
    df[col] = df[col].astype(str)
    df[col] = le.fit_transform(df[col])
```
